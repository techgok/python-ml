```python
import pandas as pd
d1={"name":["Sai","Sub","poo"],"age":[20,21,22]}
d2={"qualified":[True,True,False]}
df1=pd.DataFrame(d1)
print("DF1:\n",df1)
df2=pd.DataFrame(d2)
print("DF2:\n",df2)
newdf=df1.join(df2)
print("joining:\n",newdf)
```

```python
df1=pd.DataFrame([[11,10],[3,7]])
df2=pd.DataFrame([[11,18],[9,4]])
def myfunc(a,b):
  if(a.sum()>b.sum()):
    return a
  else:
    return b
print(df1)
print(df2)
print(df1.combine(df2,myfunc))
```

```python
import pandas as pd
d1={"name":["Sai","Sub","poo"],"age":[15,16,17]}
d2={"name":["Sai","Sub","poo"],"age":[20,21,22]}
df1=pd.DataFrame(d1)
print("DF1:\n",df1)
df2=pd.DataFrame(d2)
print("DF2:\n",df2)
newdf1=df1.merge(df2,on="name",how="left")
newdf2=df2.merge(df1,on="name",how="right")
print("Left Merge:\n",newdf1)
print("Right Merge:\n",newdf2)
```

```python
#create Dataframe
df=pd.DataFrame({"team":['A','B','C','D'],"points":[90,56,78,88],"assists":[12,34,23,44],
                "rebounds":[56,43,55,12] })
#view dataframe
print(df)
#reshape dataframe from wide format to log format
df=pd.melt(df,id_vars='team',value_vars=["points","assists","rebounds"])
#view updated dataframe
print("\nReshaped DataFrame:\n",df)
```
