```python
import pandas as pd

fruits = ["Apple", "Banana", "Cherry", "Date", "Elderberry"]
prices = [120, 40, 250, 150, 300]

fruits_df = pd.DataFrame(list(zip(fruits, prices)), columns=["Fruit", "Price"])

print("--- DataFrame Info ---")
fruits_df.info()

print("\n--- Entire DataFrame ---")
print(fruits_df.to_string())

print("\n--- Descriptive Statistics for Price ---")
print(fruits_df["Price"].describe())
```
