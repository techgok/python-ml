```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range(start='2023-01-01', periods=365)
temp_values = 20 + 10 * np.sin(np.linspace(0, 2 * np.pi, 365)) + np.random.normal(0, 2, 365)

df = pd.DataFrame({'Date': dates, 'Temperature': temp_values})
df.set_index('Date', inplace=True)

mean_temp = df['Temperature'].mean()
max_temp = df['Temperature'].max()
min_temp = df['Temperature'].min()

monthly_avg = df.resample('ME').mean()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temperature'], label='Daily Temperature', alpha=0.5)
plt.plot(monthly_avg.index, monthly_avg['Temperature'], label='Monthly Average', linewidth=3, color='red')
plt.title('Daily Temperatures Over a Year')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()

print(f"Mean: {mean_temp}")
print(f"Max: {max_temp}")
print(f"Min: {min_temp}")
print(monthly_avg)
```
