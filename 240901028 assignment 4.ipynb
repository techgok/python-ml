```python
import numpy as np

def calculate_metrics(p1, p2, m1, m2):
    point1 = np.array(p1)
    point2 = np.array(p2)
    dist = np.linalg.norm(point1 - point2)

    matrix1 = np.array(m1)
    matrix2 = np.array(m2)
    dot_prod = np.dot(matrix1, matrix2)
    
    return dist, dot_prod

p_a = (1, 2)
p_b = (4, 6)
mat_a = [[1, 2], [3, 4]]
mat_b = [[5, 6], [7, 8]]

distance, matrix_result = calculate_metrics(p_a, p_b, mat_a, mat_b)

print(distance)
print(matrix_result)
```
