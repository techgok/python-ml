```python
import pandas as pd
#assign data
data={"Name":['Sai','Yuvasri','Poovannan','Sabarish','Sangamithra','Rithika','Rubasri'],
      'Age':[20,18,19,20,18,17,19],'Gender':['M','F','M','M','F','F','F'],'Marks':[90,76,'NAN',74,65,'NAN',71]}
      #convert into dataframe
df=pd.DataFrame(data)
      #display data
print("ORiginal DataFrame:\n",df)
c=avg=0
for ele in df['Marks']:
  if str(ele).isnumeric():
    c+=1
    avg+=ele
avg/=c
#replace missing values
df=df.replace(to_replace='NAN',value=avg)
#display data
print("\nReplacing NAN with average marks:\n",df)
```

```python
#filter top scring students
df=df[df['Marks']>=75]
#remove age row
df=df.drop(['Age'],axis=1)
#display data
print("\nFiltering and Dropping func:\n",df)
```

```python
one=pd.DataFrame({'Name':['Sai','Sub','poo','Yuvi','Rithika'],
                  'Subject_id':['sub1','sub2','sub4','sub6','sub5'],
                  'Marks_scored':[98,95,96,92,88]},
                 index=[1,2,3,4,5])
two=pd.DataFrame({'Name':['Sanjay','Sangamithra','Rubasri','Nikesh','Somnath'],
                  'Subject_id':['sub2','sub4','sub3','sub6','sub5'],
                  'Marks_scored':[86,95,87,9,91]},
                 index=[1,2,3,4,5])
print("\nOriginal DataFrame 1:\n",one)
print("\nOriginal DataFrame 2:\n",two)
print(pd.concat([one,two]))
```
