```python
import pandas as ppd
dff=ppd.DataFrame()
print(dff)
```

```python
emp=ppd.Series(['Parker','Johan','Smith','william'])
id=ppd.Series([102,107,109,114])
frame={'Emp':emp,"Id":id}
result=ppd.DataFrame(frame)
print("\nSeries to Data frame\n")
print(result)
```

```python
print("\n Extracting one column:\n")
print(result['Emp'])
```

```python
print("\nAdding new column:\n")
result['Age']=ppd.Series([35,24,40,38])
print(result)
```

```python
print("\nDeleting one column:\n")
del result ['Age']
print(result)
```

```python
print("\nExtracting the third row:\n")
print(result.loc[2])
```

```python
print('\nSlice rows:\n',result[1:3])
```

```python
d2=ppd.DataFrame([['Dale',123],['Mark',143]],columns=['Emp','ID'])
print("\nAdding new row values:\n",ppd.concat([result,d2]))
```

```python
print("\nDeleting particular row:\n",result.drop(1))
```
