```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/muthuram/Downloads/flight_data.csv")

# Task a
stats = df['Departure_Delay'].describe()
print("Summary Statistics for Departure Delays:")
print(stats)
print("\n")

# Task b
avg_delay = df.groupby('Airline')['Departure_Delay'].mean()
print("Average Delay per Airline:")
print(avg_delay)
print("\n")

# Task c
common_origin = df['Origin'].mode()[0]
common_dest = df['Destination'].mode()[0]
print(f"Most Common Origin: {common_origin}")
print(f"Most Common Destination: {common_dest}")
print("\n")

# Task d
df.boxplot(column=['Departure_Delay'])
plt.title('Distribution of Departure Delays')
plt.ylabel('Minutes')
plt.show()
```
