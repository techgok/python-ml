```python
import pandas as pd

df = pd.read_csv('student_scores.csv')

# Task a
stats = df[['Maths', 'Physics', 'Chemistry']].describe()
print(stats)

# Task b
df['Average Score'] = df[['Maths', 'Physics', 'Chemistry']].mean(axis=1)
print(df[['Name', 'Average Score']])

# Task c
subjects = ['Maths', 'Physics', 'Chemistry']
below_60_count = (df[subjects] < 60).sum(axis=1)
struggling_students = df[below_60_count > 2]
print(struggling_students[['Name']])
```
